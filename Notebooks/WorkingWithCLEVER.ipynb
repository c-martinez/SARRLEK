```python
import pandas as pd
import pickle
from collections import defaultdict
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_rows', 500)
```

```python
d = defaultdict(list)
with open('Radlex&Clever/CLEVER.txt') as f:
    content = f.read().splitlines()
    f.close()  
for line in content:
    x = line.split('|')
    d[x[2]].append(x[1])
```

```python
d.keys()
```

```python
with open('Clever_dict.pickle', 'wb') as handle:
    pickle.dump(d, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
