```python
from rdflib import Graph, ConjunctiveGraph
import pandas as pd
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_rows', 500)

g = ConjunctiveGraph()
```

```python
g.parse("Radlex&Clever/owlapi.xrdf")
```

```python
len(g)
```

```python
for subj, pred, obj in g:
    if (subj, pred, obj) not in g:
        raise Exception("It better be!")
```

```python
qry = """
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX radlex: <http://www.radlex.org/RID/#>

SELECT DISTINCT ?RID ?name ?sub_name1 ?sub_name2 ?sub_name3 ?synonym ?synonym1 ?synonym2 ?synonym3
WHERE
{
 ?term radlex:Preferred_name ?name .
 ?term rdfs:label ?RID .
 FILTER regex(?RID,"RID1302",'i')
 OPTIONAL{?term radlex:Synonym ?synonym  .}
 
 OPTIONAL{?term radlex:Has_Part ?Sub_type1  .
 ?Sub_type1 radlex:Preferred_name ?sub_name1 .
 ?Sub_type1 rdfs:label ?sub1_RID .
 
 OPTIONAL{?Sub_type1 radlex:Synonym ?synonym1  .}
 
 OPTIONAL{?Sub_type1 radlex:Has_Part ?Sub_type2  .
 ?Sub_type2 radlex:Preferred_name ?sub_name2 .
 ?Sub_type2 rdfs:label ?sub2_RID .
 
 OPTIONAL{?Sub_type2 radlex:Synonym ?synonym2  .}
 
 OPTIONAL{?Sub_type2 radlex:Has_Part ?Sub_type3  .
 ?Sub_type3 radlex:Preferred_name ?sub_name3 .
 ?Sub_type3 rdfs:label ?sub3_RID .
 
 OPTIONAL{?Sub_type3 radlex:Synonym ?synonym3  .}
 

 }}}}
 """

rows  = g.query(qry)
```

```python
len(rows.bindings)
```

```python
res = [{str(k): str(v) for k, v in binding.items()} for binding in rows.bindings]
```

```python
df = pd.DataFrame(res)
    
df.head(2000)
```

```python
df_new = df.loc[df['name'] == 'right lung']
df_new
```

```python
Lst = [ df['sub_name1'].tolist(), df['sub_name2'].tolist(), df['synonym1'].tolist(), df['synonym2'].tolist()]
```

```python
print(Lst)
```

```python
updated_list = []
for l in Lst:
    for data in l:
        if str(data) == 'nan':
            pass
        else:
            updated_list.append(data)
updated_list = list(set(updated_list))
```

```python
print(updated_list)
```

```python
keys = []
for i in range(len(updated_list)):
    keys.append('right lung')
keys_df = pd.DataFrame(keys, columns=['keys'])
final_df = pd.DataFrame(updated_list, columns=['meanings'])
final_df= pd.merge(keys_df, final_df, right_index = True, left_index = True)
final_df
```

```python
final_df.to_csv("Data/Right_lung.txt", sep = "|")
```
